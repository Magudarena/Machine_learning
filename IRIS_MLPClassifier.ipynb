```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
print(iris.data)
```

```python
print(iris.target)
```

```python
data=iris.data
```

```python
target=iris.target
```

```python
data.shape
```

```python
target.shape
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(100,100))
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainX, testX, trainY, testY=train_test_split(data,target,test_size=0.2, stratify=target)
```

```python
#standaryzacja polega na (x-rednia od x)/odchylenie standardowe
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler = scaler.fit(data)
```

```python
trainX_std=scaler.transform(trainX)
testX_std=scaler.transform(testX)
```

```python
mlp.fit(trainX,trainY)
```

```python
pred=mlp.predict(testX)
```

```python
print(pred)
```

```python
print(testY)
```

```python
print((testY!=pred).sum())
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(testY,pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
sns.heatmap(confusion_matrix(testY,pred),annot=True, cmap = 'jet')
```
