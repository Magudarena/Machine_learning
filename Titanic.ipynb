```python
def get_title(name):
    title_search = re.search(' ([a-zA-Z]+)\.', name)
    if title_search:
      return title_search.group(1) #lub return title_search[1]
    return ""
```

```python
from keras import models
from keras.layers import Dense,Dropout
from keras import regularizers
```

```python
def titanic():
    model=models.Sequential()
    model.add(Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
    model.add(Dropout(0.25)) #powoduje, ≈ºe co 4 cykle neuron zostaje usuwany
    model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001))) #coraz mniej neuron√≥w
    model.add(Dropout(0.25))
    model.add(Dense(16, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
    model.add(Dropout(0.25))
    model.add(Dense(1, activation='sigmoid')) #sigmoid do binarnych warto≈õci - zero albo jeden
    return model
```

```python
model=titanic()
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy']) #rmsprop jest oparty na propagacji wstecznej
```

```python
trainX = train_X.astype('float32')
testX = test_X.astype('float32')
```

```python
#zbi√≥r walidacyjny dostarcza informacjƒô zwrotnƒÖ czy dobrze siƒô nauczy≈Ç
history=model.fit(trainX, train_Y, epochs=30, batch_size=64, validation_split=0.1)
```
